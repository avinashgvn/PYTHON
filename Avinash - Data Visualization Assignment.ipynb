```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
from sklearn import decomposition 
from sklearn import datasets
```

```python
dataset=sns.load_dataset('iris')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,4].values

#label encoding the Target values
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_y = LabelEncoder()
y = labelencoder_y.fit_transform(y)


#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)

#Reducing the dimensions for 3D
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X = pca.fit_transform(X)


#Visualising the 3D chart
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(1, figsize=(10, 5))
plt.clf()
ax = Axes3D(fig)
ax = fig.gca(projection='3d')
# Reorder the labels to have colors matching the cluster results
ax.scatter(X[:, 0], X[:, 1], X[:, 2],label='Data Visualization using 3D',alpha=.75)
ax.legend()
ax.set_xlabel('Setosa')
ax.set_ylabel('Versicolour')
ax.set_zlabel('Virginica')
plt.show()
```
