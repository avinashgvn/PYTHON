1. Create a pie chart presenting the male/female proportion

```python
url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
titanic=pd.read_csv(url)
c=titanic['sex'].value_counts()
```

```python
plt.pie(c.values.tolist(), labels=c.index.tolist(),autopct="%1.2f%%",shadow=True)
plt.legend()
plt.show()
```

2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
titanic=pd.read_csv(url)
df=titanic[['fare','age','sex']]
df['age'].fillna(df['age'].mean(),inplace=True)
df['fare'].fillna(df['fare'].mean(),inplace=True)
```

```python
df['sex'].replace('female',0,inplace=True)
df['sex'].replace('male',1,inplace=True)
df[['sex']].dropna(axis=0)
df[['fare']].dropna(axis=0)
df.dropna()
```

```python
ax=df.plot.scatter(x='age',y='fare',c='sex',legend=True,title="Scatterplot with the Fare paid and the Age",colormap='viridis',figsize=(10,5))
ax.set_xlabel("Age")
ax.set_ylabel("Fare Paid")
```
